```python
# break- exit from the loop/current execution
for i in range(20,1,-2):
  if i == 12:
    break
  else:
    print(i)

print('Hello')
print(100)
```

```python
# add numbers in an empty list
num_list = []
for i in range(20,1,-2):
  if i == 12:
    break
  else:
    num_list.append(i)
print('Final list:',num_list)
print('Hello')
print(100)
```

```python
nms = ['sharduli','vinay','sourabh']
for nm in enumerate(nms):  # when we are using a for loop hence we dont require typecasting
  print(nm)
```

```python
nms = ['sharduli','vinay','sourabh']
for index,nm in enumerate(nms):  # when we are using a for loop hence we dont require typecasting
  print(index,nm)
```

```python
# continue- same as that of break but it continues further execution
g = [1,2,3,4,5,1,6,7,1,8,9,1]
for value in g:
  if value == 1:
    continue  #main purpose is to skip the eelemnt in condition
  else:
    print(value)
```
