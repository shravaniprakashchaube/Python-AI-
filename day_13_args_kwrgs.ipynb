**Q. What are different types of arguments in python?**

```python
#1. Postional argument- position/sequence matters in this case

def info(name,age,salary):
  print('Name:',name,'Age:',age,'Salary:',salary)


info('amol',20,34000)
```

```python
# if we change the sequence in calling
info(21,45000,'vinay',)
```

```python
#2. Keyword args: in which sequence order doesnt matter
# bcz we provide key=value assignment

info(age='21',salary=45000,name='vinay',)
```

```python
#3 Default args- in declaration we add default values
# when user is not supplying the values throgh a calling of a function then it will take default values
# if user supplies the value then it takes those

def test(name='Guest',place = 'Pune'):
  print(name,place)

test()
```

```python
test('Sujit','Mumbai')
```

```python
#4. Variable length args
#4.1 Var. len. Positional args/*args
#42. Var. len. Keyword args/kwrgs
```

```python
def test(*a):
  print(a)

test(10)
test(4,5,6,'python',[1,2,3])
test()
```

```python
#4.2 Var. len keyword: **kwrgs
# we must need to supply key=value pairs
def test(**a):
  print(a)

test()
```

```python
test(a=10,b=80)
test(name='python',age=35)
```
