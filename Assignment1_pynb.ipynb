**Case Conversion**


*  capitalize(): Converts the first character to uppercase and the rest to     lowercase.
* lower(): Converts all characters in the string to lowercase.
*   upper(): Converts all characters in the string to uppercase
*  title(): Converts the first character of each word to uppercase.
*   swapcase(): Swaps cases, making lowercase letters uppercase and vice versa.











```python
txt = "hello"

x = txt.capitalize()

print (x)
```

```python
txt = "Hello "

x = txt.lower()

print(x)
```

```python
txt = "Hello "

x = txt.upper()

print(x)
```

```python
txt = "hello"

x = txt.title()

print(x)
```

```python
txt = "Hello "

x = txt.swapcase()

print(x)
```

**Searching and Finding**
Methods like count(), find(), and index() are used to locate substrings within a string. find() returns the starting index of the first match or -1, while index() does the same but raises an error if not found. startswith() and endswith() check if a string begins or ends with a specific value, returning True or False.

```python
txt = "Hello, welcome to my world."

x = txt.find("welcome")

print(x)
```

```python
txt = "I love apples, apple are my favorite fruit"

x = txt.count("apple")

print(x)
```

```python
txt = "Hello, welcome to my world."

x = txt.index("welcome")

print(x)
```

```python

#The endswith() method returns True if the string ends with the specified value, otherwise False.
txt = "Hello, welcome to my world."

x = txt.endswith(".")

print(x)
```

```python

txt = "Hello, welcome to my world."

x = txt.startswith("Hello")

print(x)
```

**Splitting and Joining**
The split() method divides a string into a list based on a separator. Conversely, join() combines items from an iterable into a string, using the original string as the separator.

```python
txt = "welcome to the jungle"

x = txt.split()
print(x)
```

```python
myTuple = ("John", "Peter", "Vicky")

x = '#'.join(myTuple)

print(x)
```

**Formatting and Trimming**
Strings can be trimmed using strip(), lstrip(), and rstrip() to remove leading/trailing whitespace or specified characters. The replace() method substitutes occurrences of a substring with another. format() and f-strings are used for embedding values within strings.

```python
txt = "     banana     "

x = txt.strip()

print("of all fruits", x, "is my favorite")
```

```python
txt = "     banana     "

x = txt.lstrip()

print("of all fruits", x, "is my favorite")
```

```python
txt = "     banana     "

x = txt.rstrip()

print("of all fruits", x, "is my favorite")
```

```python
txt = "I like bananas"

x = txt.replace("bananas", "apples")

print(x)
```

**Character Type Testing**
These methods check the character content of a string. Examples include isalnum() (alphanumeric), isalpha() (alphabetic), isdigit() (digits), and isspace() (whitespace). isupper() and islower() check the case of characters. These methods return boolean values.

```python
txt = "Company12"

x = txt.isalnum()

print(x)
```

```python
txt = "CompanyX"

x = txt.isalpha()

print(x)
```

```python
txt = "   "

x = txt.isspace()

print(x)
```

```python
txt = "THIS IS NOW!"

x = txt.isupper()

print(x)
```

```python
txt = "50800"

x = txt.isdigit()

print(x)
```
