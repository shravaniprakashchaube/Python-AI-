```python
# declaration
def addition():
  num1 = int(input('Enter num1:'))  # default data type str
  num2 = int(input('Enter num2:'))
  print('Addition is:',num1+num2)


# calling
addition()
```

```python
# take the input outside the function
# declaration
def addition(num1,num2):
  print('Addition is:',num1+num2)

num1 = int(input('Enter num1:'))  # default data type str
num2 = int(input('Enter num2:'))
# calling
addition(num1,num2)
```
