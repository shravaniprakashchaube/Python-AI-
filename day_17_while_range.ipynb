```python
pin = int(input('Enter the pin:'))
while pin != 1234:
  print('Login Failed')
  pin = int(input('Enter the pin:'))
else:
  print('Login Succesfull')
```

# **Transfer statements**

```python
# contains 3 options
# 1. break
# it is used to break the loop
# range(10) --> start from 0 and stop at num-1= 9
range(10)
```

```python
list(range(10))
```

```python
# we can also give start and stop
list(range(3,15))
```

```python
# we can also give start and stop  and step
list(range(3,15,1)) # default step 1
```

```python
list(range(3,15,2)) # default step 1
```

```python
# if num is 9 then stop
for num in range(3,15,2):
  #print(num)
  if num == 9:
    break
  else:
    print(num,end = ' ')
```
