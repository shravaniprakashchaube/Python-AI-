```python
# Tuple is Immutable in nature
#----------------------------------------------------
# In data type we have 2 approaches
# 1. Mutable data types- in which we can directly perform the changes(add,update,remove,insert,del)
# Ex. List, Set, Dict
# 2. Immutable data types- in which we can't perform the change in the original object, change remain temp.
# Ex. Tuple,string, range,
```

```python
p = [10,20,30]
p
```

```python
# list is mutable
# lets try to change last eleement 30
p[-1] = 'Akshay'
```

```python
p
```

```python
# remove 10
p.remove(10)
```

```python
p
```

```python
# use insert(index,value) to add an element using index
p.insert(0,'LAXMI')
```

```python
nums
```

```python
# Unpacking- It is a process of separating the individual object ans store it in separate variable called as Unpacking
num1,name,num2 = nums
```
