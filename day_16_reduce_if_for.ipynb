```python
# reduce(function,iterable)
# when usr wants to reduce a sequence as a single entity/object
a = [6,1,3,2,5]

# reduce is present in a module(file) functools
# we need to import reduce from functools module
from functools import reduce
```

```python
reduce(lambda x,y:x+y,a)
```

```python
# check multiple conditions using elif
pin = int(input('Enter your pin:'))
if len(str(pin)) == 4:
  if pin == 4455:
    print('Welcome to SBI Bank')
  elif pin== 1234:
    print('Welcome ICICI bank')
  elif pin == 9090:
    print('Welcome to Kotak bank')
  else:
    print('Wrong pin entered...')
else:
  print('Pin must be 4 digit')
```
