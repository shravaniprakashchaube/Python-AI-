```python
# Features of a dict
# duplicate keys are not allowed
g = {1:10,1:1000,1:'java'}
g
```

```python
# Rule- If we have multiple values with same key present in the dict then it takes last recent pair
# in above case 1:'java' is last key value pair
```

```python
# Iterable: collection of elements, we have more than oe element
s = {(1,2),'sourabh'}
s
```

```python
h = {1:10,[2,4]:1000,3:'java'}  # here we are trying to set a list as a key which is not allowd- check below rule
h
```

```python
# in a set we can not give/add MUTABLE objects
# & in dict we can not set key as a mutable object
```

```python
# Dict is Mutable
```

```python
# create an empty dict
x = {}
x
```

```python
# add pair/s- use update(dict)
x.update({1:10,2:20})
```

```python
x
```

```python
# we can also use update to update already existing value of concerned key
x.update({2:'sql'})
x
```

```python
# we can aslo use setdefault method
help(x.setdefault)
```

```python
x.setdefault(4)  # for key 4 it will add None as a value 4: None
```

```python
x
```

```python
# instead to None we want another value
x.setdefault(5,'ABC')
x
```

```python
# if key is already present n we  try to give a new value or if we keep it empty
x.setdefault(5)  # if key already exist in the dict, then setdefault returns its value, it wont update it to None
```
