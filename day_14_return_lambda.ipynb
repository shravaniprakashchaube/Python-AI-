```python
# Output from One function will supply as an input to other function

def exam(per):
  # lets add condition inside
  if per>= 60:
    return 'pass'
  else:
    return 'fail'

result = exam(87)

def final(result):
  if result=='pass':
    print('Welcome for interview process')
  else:
    print('Please try next time')

final(result)
```

```python
a = exam(75)
final(a)
```
