```python
# List comprehension- to create a new list using short code snippet
# it always one line code
# syntax: [var+operation  for var in iterable]
```

```python
# I want a list of square of 1-10  numbers
[var**2 for var in range(1,11)] # ** is an exponetial operator - power of
```

```python
nms = ['nikhil','laxmi','vinay']
[nm.upper() for nm in nms]
```

```python
nms = ['nikhil','laxmi','vinay']
# we can also pass a condition after for loop
[nm.upper() for nm in nms if len(nm)<6]
```

```python
list(filter(lambda nm:len(nm)<6,nms))
```

```python
z = [100,3,5,2,10,7]
# -1 for even number and +1 for odd number

[n-1 if n%2==0 else n+1 for n in z]
```

```python
n = 'shridhar'
# [(index,char)]
[val for val in enumerate(n)]
```

```python
n = 'shri d  h     ar'
# [(char,index)]
[(val,index) for index,val in enumerate(n) if val!=' ']
```

```python
# dict comprehension: (key,value)
n = 'shri d  h     ar'
# {index:value}
{index:val for index,val in enumerate(n) if val!=' '}
```
