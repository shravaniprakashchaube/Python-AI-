# **Higher order function**
- a function which needs another function as an input
- syntx: function(function)
- We have 3 higher order functions
1. map(function,iterable)
2. filter(function,iterable)
3. reduce(function,iterable)

```python
# 1. map - perform an operation present/given in a function over the each element  from an iterable
k = [10,20,30,40]

# lets multiply each number by 2
map(lambda num: num*2,k)
```

```python
tuple(map(lambda num: num*2,k))
```

```python
# put a set inside
tuple(map(lambda num: num*2,{1,1,1,2,2,2,4,5,3}))
```

```python
dict(((1,2),(3,4)))
```

```python
k = [10,20,30,40]

# lets multiply each number by 2
mult = lambda num: num*2

list(map(mult,k))
```

```python
# we can also use normal function
a = [4,5,6,7]  # iterable

def test(num): # function
  return num*100

tuple(map(test,a))
```

```python
# label even numbers as 'even' and odd numbers as 'odd'
a = [4,5,6,7]  # iterable

def test(num): # function
  if num%2==0:
    return 'even'
  else:
    return 'odd'

tuple(map(test,a))
```

```python
#2. filter(function,iterable)
# filter is condition based
# it returns only those values for which condition result it True
nums = (10,-2,5,-11,50,-7)

# filter -ve numbers
list(filter(lambda n:n<0,nums))
```

```python
list(filter(lambda n:n%5==0,nums))
```
